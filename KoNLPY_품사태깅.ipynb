# 문자 전처리

```python
# 정규표현식으로 문자 전처리 하기
import re

def preprocessing(text):
  # 한글, 영문, 숫자만 남기고 모두 제거하도록 함.
  text = re.sub('[^ㄱ-ㅎ|가-힣|a-z|A-Z|0-9|]', ' ', text)
  # 중복으로 생성된 공백값을 제거
  text = re.sub('[\s]+', '', text)
  # 영문자를 소문자로 만듬
  text = text.lower()
  return text
```

```python
# 결측값 
re_train = train['title'].dropna("")
re_test = test['title'].dropna("")
re_train
re_test
```

```python
re_train = re_train.map(preprocessing)
re_test = re_test.map(preprocessing)
re_train
re_test
```

```python
%time Okt().pos("버스의 운행시간을 문의했었다. 어?!", stem=True)
```

```python
# 형태소 분석기(Okt) 불러오기
# ['Josa', 'Eomi', 'Punctuation']
okt_pos =  Okt().pos("버스의 운행시간을 문의했었다. 어?!", stem=True)

clean_text = list()

for op in okt_pos:
  if op[1] not in ["Josa", "Eomi", "Punctuation"]:
    clean_text.append(op[0])    
    # print(op[0], op[1])    
    
" ".join(clean_text)
```

```python
# 형태소 분석기(Okt) 불러오기
# ["Josa", "Eomi", "Punctuation"] 조사, 어미, 구두점 제거
def okt_clean(text):
  okt_pos =  Okt().pos(text, stem=True)
  clean_text = list()
  for op in okt_pos:
    if op[1] not in ["Josa", "Eomi", "Punctuation"]:
      clean_text.append(op[0])    
    # print(op[0], op[1])        
  return " ".join(clean_text)

okt_clean("기차 운행시간을 문의했었습니다. 어?!")
```

```python
!pip install tqdm --upgrade
```

```python
from tqdm import tqdm
tqdm.pandas()

clean_re_train = re_train.progress_map(okt_clean)
clean_re_test = re_test.progress_map(okt_clean)
```

```python
# 불용어 제거
def remove_stopwords(text):
  token = text.split(' ')
  stop = ['합니다', '하는', '할', '하고', '한다', '그리고', '입니다', '그', '등', '이런', '및', '제', '더']
  meaningful_words = [w for w in token if not w in stop]
  return ' '.join(meaningful_words)

remove_train = clean_re_train.map(remove_stopwords)
remove_test = clean_re_test.map(remove_stopwords)
remove_train
remove_test
```
